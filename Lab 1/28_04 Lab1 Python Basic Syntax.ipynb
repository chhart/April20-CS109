## Functions

```python
def square(x):
    return(x*x)

def cube(x):
    return(x*x*x)

print(square(5))
print(cube(5))
```

Functions are 'first-class', i.e. they can be passed to one another

```python
def sum_of_anything(x,y,f):
    print(x,y,f)
    return(f(x)+f(y))
sum_of_anything(3,4,square)
```

Python functions can have positional arguments and keyword arguments:
* positional are stored in a tuple
* keyword arguments in a dictionary

```python
def f(a,b,*posargs,**dictargs):
    print("got",a,b,posargs, dictargs)
    return a
print(f(1,3))
print(f(1,3,4,d=1,c=2))
```

We can break out of a loop based on a condition - below is a 'for' loop:

```python
for i in range(10):
    print(i)
    if (i > 5):
        break
```

'while' loops are also supported

'continue' brings us to the next iteration of the loop and skips the code below, whilst 'break' breaks out of it

```python
i=0
while i < 10:
    print(i)
    i=i+1
    if i<5:
        continue
    else:
        break
```

# And all together!

```python
# convert all of the hamlet tokens into lower case with list comprehension

hamletlctokens=[word.lower() for word in hamlettokens]
hamletlctokens.count("thou")
```

We're going to try to get a dictionary with each word and the number of times it appears in Hamlet

```python
uniquelctokens = set(hamletlctokens) # set is kinda like 'SELECT DISTINCT'

tokendict = {}
for ut in uniquelctokens:
    tokendict[ut] = hamletlctokens.count(ut)
```

```python
tokendict.items()
```

```python
# let's try to get the top 100
# sorting a dictonary by values isn't easy... 

sorted_tokens = sorted(tokendict.items(), reverse = True, key=lambda x: x[1])
sorted_tokens[:100]
```

```python
topfreq = sorted_tokens[:20]
print(topfreq)
```

```python
pos = np.arange(len(topfreq)) # this gives us the x coordinates of the bars
plt.bar(pos, [e[1] for e in topfreq])
```

```python
# we need to add the labels for the x axis
plt.bar(pos, [e[1] for e in topfreq]);
plt.xticks(pos+0.4, [e[0] for e in topfreq]);
```
