### Filtering

Two ways to filter a dataframe:

```python
df.query("rating > 4.5")
```

Or we could create a mask and then use it to index the rows we want

```python
df[df.year < 0]
```

We can combine these with a boolean '&'

```python
df[(df.year < 0) & (df.rating > 4)] # a list of ancient books with high ratings
```
